```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt

image_path = r"C:\Users\govin\Downloads\fr1.png" 
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

detector = MTCNN()

detections = detector.detect_faces(image)

def draw_faces(image, detections):
    for detection in detections:
        
        x, y, width, height = detection['box']
        cv2.rectangle(image, (x, y), (x + width, y + height), (255, 0, 0), 2)
        
    
        for key, point in detection['keypoints'].items():
            cv2.circle(image, point, 2, (0, 255, 0), 2)
    return image


image_with_detections = draw_faces(image.copy(), detections)


plt.figure(figsize=(10, 10))
plt.imshow(image_with_detections)
plt.axis('off')
plt.show()
```

```python
from mtcnn import MTCNN
import cv2
import matplotlib.pyplot as plt

image_path = r"C:\Users\govin\Downloads\fr3.jpg"  
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

detector = MTCNN()


detections = detector.detect_faces(image)


def draw_faces(image, detections):
    for idx, detection in enumerate(detections):
        
        x, y, width, height = detection['box']
        confidence = detection['confidence']
        
        
        cv2.rectangle(image, (x, y), (x + width, y + height), (255, 0, 0), 2)
        
        
        label = f"Face {idx+1} ({confidence*100:.2f}%)"
        font_scale = 0.3  
        font_thickness = 1 
        cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 0, 0), font_thickness)
        
        
        for key, point in detection['keypoints'].items():
            cv2.circle(image, point, 2, (0, 255, 0), 2)
    
    return image


if len(detections) > 0:
    print(f"Detected {len(detections)} faces.")
else:
    print("No faces detected.")


image_with_detections = draw_faces(image.copy(), detections)


plt.figure(figsize=(10, 10))
plt.imshow(image_with_detections)
plt.axis('off')
plt.show()
```
